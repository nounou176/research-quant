# **Phần III. Trực quan hóa dữ liệu**

```python
# Hàm set_index() được sử dụng để đặt chỉ mục DataFrame cột thời gian
df=df.reset_index()
df=df.set_index('time')
df
```

```python
# Trực quan hóa dữ liệu cổ phiếu
from vnstock import *
from vnstock.chart import *
import plotly.subplots as ms
import plotly.graph_objects as go

fig = ms.make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.02)

fig.add_trace(go.Candlestick(x=df.index,
                             open=df['open'],
                             high=df['high'],
                             low=df['low'],
                             close=df['close'],
                             name='Candlestick'),
              row=1, col=1)

# Thêm biểu đồ Volume dưới dạng cột
fig.add_trace(go.Bar(x=df.index, y=df['volume'], name='Volume'),
              row=2, col=1)

# Cập nhật bố cục của biểu đồ
fig.update_layout(title='MWG Candlestick Chart',
                  yaxis_title='Price',
                  xaxis_title='Date',
                  xaxis_rangeslider_visible=False,
                  bargap=0.3,)  # Khoảng cách giữa các cột Volume

# Hiển thị biểu đồ
```
